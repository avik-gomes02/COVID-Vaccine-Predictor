```python
from sklearn import linear_model
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 

from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC

from pandas import ExcelWriter
from pandas import ExcelFile
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_excel('sample.xlsx', sheet_name = 'Sheet1')

df = df.dropna(axis='columns')
df
```

```python
X = df.loc[:, df.columns != 'NUM_SYMPTOMS']
Y = df['NUM_SYMPTOMS']
```

```python
#splitting data for training and testing
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size = 0.3, random_state = 6) #30% of data for testing

cVals = [0.0001, 0.001, 0.01, 0.1, 1, 10]
  
acc_train_svm_linear = []
acc_test_svm_linear = []
c_svm_linear = []  
    
def model_linear(c):
    # Creating the linear SVM model
    model_linear = OneVsRestClassifier(SVC(kernel = 'linear', C = c)) 
    
    # Fitting the model with training data
    model_linear.fit(X_train, y_train)

    # TODO - Find the prediction and accuracy on the training set - 5 points
    Yhat_svc_linear_train = model_linear.predict(X_train)
    acc_train = model_linear.score(X_train, y_train)
    
    # Adding testing accuracy to acc_train_svm
    acc_train_svm_linear.append(acc_train)
    print('Train Accuracy = {0:f}'.format(acc_train))
    
    # TODO - Find the prediction and accuracy on the test set - 5 points
    Yhat_svc_linear_test = model_linear.predict(X_test)
    acc_test = model_linear.score(X_test, y_test)
    
    # Adding testing accuracy to acc_test_svm
    acc_test_svm_linear.append(acc_test)
    print('Test Accuracy = {0:f}'.format(acc_test))
    
    # Appending value of c for graphing purposes
    c_svm_linear.append(c)
```

```python
for c in cVals:
    model_linear(c)

plt.plot(c_svm_linear, acc_train_svm_linear, 'o')
plt.plot(c_svm_linear, acc_test_svm_linear, 'o')
plt.xlabel('c')
plt.ylabel('Accuracy')
plt.grid()

# Use the following function to have a legend
plt.legend(['Training Accuracy', 'Test Accuracy'], loc='upper right')
```

```python
acc_train_svm_rbf = []
acc_test_svm_rbf = []
c_svm_rbf = []

def model_rbf(c):
    model_rbf = OneVsRestClassifier(SVC(C = c))
    
    model_rbf.fit(X_train, y_train)

    Yhat_svc_rbf_train = model_rbf.predict(X_train)
    acc_train = model_rbf.score(X_train, y_train)

    acc_train_svm_rbf.append(acc_train)
    print('Train Accuracy = {0:f}'.format(acc_train))

    Yhat_svc_rbf_test = model_rbf.predict(X_test)
    acc_test = model_rbf.score(X_test, y_test)
    
    acc_test_svm_rbf.append(acc_test)
    print('Test Accuracy = {0:f}'.format(acc_test))

    c_svm_rbf.append(c)
  
```

```python
for c in cVals:
    model_rbf(c)

plt.plot(c_svm_rbf, acc_train_svm_rbf, 'o')
plt.plot(c_svm_rbf, acc_test_svm_rbf, 'o')
plt.xlabel('c')
plt.ylabel('Accuracy')
plt.grid()

# Use the following function to have a legend
plt.legend(['Training Accuracy', 'Test Accuracy'], loc='lower right')
```

```python
acc_train_svm_poly = []
acc_test_svm_poly = []
c_svm_poly = []

def model_poly(c):
    model_poly = OneVsRestClassifier(SVC(kernel = 'poly', C = c))
    
    model_poly.fit(X_train, y_train)
    
    # TODO - Find the prediction and accuracy on the training set - 5 points
    Yhat_svc_poly_train = model_poly.predict(X_train)
    acc_train = model_poly.score(X_train, y_train)
    
    # Adding testing accuracy to acc_train_svm
    acc_train_svm_poly.append(acc_train)
    print('Train Accuracy = {0:f}'.format(acc_train))
    
    # TODO - Find the prediction and accuracy on the test set - 5 points
    Yhat_svc_poly_test = model_poly.predict(X_test)
    acc_test = model_poly.score(X_test, y_test)
    
    # Adding testing accuracy to acc_test_svm
    acc_test_svm_poly.append(acc_test)
    print('Test Accuracy = {0:f}'.format(acc_test))
    
    # Appending value of c for graphing purposes
    c_svm_poly.append(c)
```

```python
cVals = np.geomspace(0.000001, 0.001, 5)
for c in cVals:
    model_poly(c)

plt.plot(c_svm_poly, acc_train_svm_poly, 'o')
plt.plot(c_svm_poly, acc_test_svm_poly, 'o')
plt.xlabel('c')
plt.ylabel('Accuracy')
plt.grid()

# Use the following function to have a legend
plt.legend(['Training Accuracy', 'Test Accuracy'], loc='upper right')
```
